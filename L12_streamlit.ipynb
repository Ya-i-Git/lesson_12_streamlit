# Streamlit

First you need to install streamlit and matplotlib

Обязательно добавляем requarmentes.txt

## main.py

Для запуска скрипта в консоле надо будет прописать **streamlit run main.py**

st.session_state - Записывает в себя состояние переменной. Это нужно чтобы изменять значение при нажатии на кнопку. Пока что как я понял это надо делать чтобы менять True на False

И при соблюдении условия будет выводиться или нет то что надо после нажатия кнопки.

Он как то прогонят код и работает в режиме ротслеживания

Для добавления нескольких страниц делаем папку с названием "pages" для порядкового номера обязательно добавляем номер "1_" в название

есть два способа отображения DataFrame: st.write() и st.dataframe()  
st.write() — универсальная функция для вывода различных объектов; для DataFrame она вызывает st.dataframe() по умолчанию, но с некоторыми отличиями.

st.dataframe() — специализированная функция для интерактивных таблиц.
